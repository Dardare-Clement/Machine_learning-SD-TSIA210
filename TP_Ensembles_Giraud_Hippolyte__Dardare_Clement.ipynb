Use the incremental evaluation of the partitions seen in class : the complexity should be  $O(ndc)$ instead of the naive $O(n^2 dc)$ version. We assume that the sorting operations such as ```argsort``` are free because we only need to run them once. Iterate in the 2-dimensions for every possible split, evaluate the quality of each split using an incremental version of the Gini index (defined next) and store the best split. (Note : a non-incremental version will be graded with half the points)
- Implement the ```gini``` method : Implement the Gini index coefficient for the case in which there are only 2 classes. In class, you saw the unweighted case: we quickly recall the unweighted then weighted version. Let $C$ be the number of different classes, $p_k(S)$ be the ratio of datapoints of class $k$ in region $S$. Then, the Gini index $G(S)$ is $$ G(S) = 1 - \sum_{k=1}^{C} p_k(S)^2$$ Given a split in which we have left and right regions $S_r$, $S_l$, let $N_r$ (resp. $N_l$) be the number of datapoints in $S_r$ (resp. $S_l$). The Gini index of the split is the combination of the Gini index of both regions: $$ \frac{N_r}{N_r + N_l}G(S_r) + \frac{N_l}{N_r + N_l}G(S_l) $$ For generalizing to the weighted case, let $w_k(S)$ be the sum of the weights of all data-points of class $k$ in $S$. The Gini index is defined as follows: $$ G(S) = 1 - \sum_{k=1}^{C} \left( \frac{w_k(S)}{  \sum_{m=1}^{C} w_m(S)} \right)^2 $$ Given a split in which we have left and right regions $S_r$, $S_l$, let $W_r = \sum_{k=1}^{C} w_k(S)$ (resp. $W_l$) the total weight on $S_r$ (resp. $S_l$). The Gini index of the partition is the combination of the Gini index of both regions: $$ \frac{W_r}{W_r + W_l}G(S_r) + \frac{W_l}{W_r + W_l}G(S_l) $$

````python
from sklearn import tree

class My_stump:
    def __init__(self):
        self.best_dimension = None
        self.best_threshold = None
        self.class_bellow_threshold = None
        self.class_above_threshold = None
        self.best_ev = None
        
    def get_threshold(self): return self.best_threshold
  
    def get_best(self): return self.best_dimension,  self.best_threshold, self.best_ev
    
    def gini_impurity(self, X0, X1, sample_weight=None): 
        
        if sample_weight is None:
            sample_weight = np.zeros(len(X))
            for i in range(len(sample_weight)):
                sample_weight[i] = 1/len(sample_weight)

           
        def weight1(a):
            c = 0
            for i in range(len(a)):
                if (a[i] == 1):
                    c += sample_weight[i]
            return c
        
        def weight0(a):
            c = 0
            for i in range(len(a)):
                if (a[i] == -1):
                    c += sample_weight[i]
            return c
        
        n0 = len(X0)
        n1 = len(X1)
        
        weight_one0 = weight1(X0)
        weight_zero0 = weight0(X0)
        weight_one1 = weight1(X1)
        weight_zero1 = weight0(X1)
        
        
        if (n0 == 0): 
            G0 = 1
        else:
            G0 = 1 - (weight_zero0 /(weight_zero0 + weight_one0))**2 - (weight_one0/(weight_zero0 + weight_one0))**2
            
        if (n1 == 0):
            G1 = 1
        else:
            G1 = 1 - (weight_zero1 /(weight_zero1 + weight_one1))**2 - (weight_one1/(weight_zero1+ weight_one1))**2
        
        
        G = (n0/(n1 + n0)) * G0 + (n1/(n0 + n1)) * G1
        
        return G
    
    def fit(self, X, y, sample_weight=None):
        
        if sample_weight is None:
            sample_weight = np.zeros(len(X))
            for i in range(len(sample_weight)):
                sample_weight[i] = 1/len(sample_weight)
        

        point_abs = []
        point_ord = []
        
        for i in range(len(X)):
            point_abs.append([X[i][0], y[i], sample_weight[i]])
            point_ord.append([X[i][1], y[i], sample_weight[i]])
            
        def one_dim_tresh(L):

            abs_sorted = np.array(sorted(L, key = lambda x : x[0]))
            results = [[abs_sorted[0][0] - 1, self.gini_impurity([], y, sample_weight)]] # will store all treshold assocoated to the gini index, initialised

            for i in range(len(abs_sorted) - 1):
                tresh = (abs_sorted[i][0] + abs_sorted[i+1][0])/2
                G = self.gini_impurity((abs_sorted[:,1])[:i], (abs_sorted[:,1])[i:], sample_weight)
                results.append([tresh, G])
                
            results.append([abs_sorted[0][-1] + 1, self.gini_impurity(y, [])])
            tresh =  min(results, key=lambda x: x[1])
            return tresh
        
        tresh1 = one_dim_tresh(point_abs)
        tresh2 = one_dim_tresh(point_ord)
        
        compteur_below_0 = 0
        compteur_below_1 = 0
        compteur_above_0 = 0
        compteur_above_1 = 0
                       
        if (tresh1[1] < tresh2[1]):
            self.best_dimension = 0
            self.best_threshold = tresh1[0]
            
            for i in range(len(X)):
                if (X[i][0] < self.best_threshold):
                    if (y[i] == 1):
                        compteur_below_1 += 1
                    else:
                        compteur_below_0 += 1
                else:
                    if (y[i] == 1):
                        compteur_above_1 += 1
                    else:
                        compteur_above_0 += 1
                    
        else:
            self.best_dimension = 1
            self.best_threshold = tresh2[1]
            
            for i in range(len(X)):
                if (X[i][1] < self.best_threshold):
                    if (y[i] == 1):
                        compteur_below_1 += 1
                    else:
                        compteur_below_0 += 1
                else:
                    if (y[i] == 1):
                        compteur_above_1 += 1
                    else:
                        compteur_above_0 += 1
            
            
        if (compteur_below_0 < compteur_below_1):
            self.class_bellow_threshold = 0
        else:
            self.class_bellow_threshold = 1
            
        if (compteur_above_0 < compteur_above_1):
            self.class_above_threshold = 0
        else:
            self.class_above_threshold = 1
        
    def predict(self, X) :
        res = []
        for i in range(len(X)):
            if(self.best_dimension == 0):
                if(X[i][0] < self.best_threshold):
                    res.append(-1)
                else:
                    res.append(1)
            else:
                if(X[i][1] < self.best_threshold):
                    res.append(-1)
                else:
                    res.append(1)
           
        return np.array(res)
````
